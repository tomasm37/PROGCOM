```python
import random
import heapq as hq


class Personaje:
    """
    Representa un personaje mitológico latinoamericano con atributos como nombre, origen y descripción.
    """
    def __init__(self, nombre, origen, descripcion, permiso):
        self._nombre = nombre
        self._origen = origen
        self._descripcion = descripcion
        self._permiso = permiso

    # Encapsulamiento: Métodos getter para acceder a los atributos privados.
    def get_nombre(self):
        return self._nombre

    def get_origen(self):
        return self._origen

    def get_descripcion(self):
        return self._descripcion

    def tiene_permiso(self):
        return self._permiso


class Juego:
    """
    Gestiona la lógica del juego, incluyendo la selección de personajes, la interacción con el jugador y el fin del juego.
    """

    def __init__(self):
        self.personajes = [
            Personaje("Ahuizotl", "México", "Criatura acuática con manos con garras.", True),
            Personaje("La Llorona", "México", "Mujer que llora por sus hijos.", True),
            Personaje("Chupacabra", "Puerto Rico", "Criatura que chupa la sangre de animales.", True),
            Personaje("Pombero", "Argentina", "Ser de los bosques.", True),
            Personaje("Cacique", "Colombia", "Entidad maligna de los ríos.", True),
            Personaje("Siguanaba", "Honduras", "Criatura que se aparece a los hombres infieles.", True),
            Personaje("Duendes", "Perú", "Criatura pequeña que vive en las montañas.", True),
            Personaje("Nahual", "Guatemala", "Persona con la capacidad de transformarse en animales.", True),
            Personaje("Yacumama", "Perú", "Seres mitológicos que habitan en ríos y lagunas.", True),
            Personaje("Mapinguari", "Brasil", "Criatura similar a un oso perezoso.", True),
        ]
        self.personajes_seleccionados = []
        self.monstruo_sin_permiso = None  # Variable para guardar al monstruo sin permiso

    def seleccionar_personajes(self):
        """Selecciona 6 personajes aleatoriamente, uno de ellos sin permiso."""
        self.personajes_seleccionados = random.sample(self.personajes, 6)
        # Elegir un personaje al azar para no tener permiso
        self.monstruo_sin_permiso = random.choice(self.personajes_seleccionados)
        self.monstruo_sin_permiso._permiso = False

        # Reemplazar la descripción del monstruo sin permiso por uno no utilizado
        personajes_no_utilizados = [p for p in self.personajes if p not in self.personajes_seleccionados]
        if personajes_no_utilizados:
            reemplazo = random.choice(personajes_no_utilizados)
            self.monstruo_sin_permiso._descripcion = reemplazo.get_descripcion()


    def mostrar_personajes(self):
        """Muestra la lista de nombres de personajes a cotejar."""
        for i, personaje in enumerate(self.personajes_seleccionados):
            print(f"{i + 1}. {personaje.get_nombre()}")

    def cotejar_personaje(self, personaje):
        """Permite al usuario verificar los datos del personaje y llamar a los cazafantasmas si cree que es falso."""
        print(f"\nCotejando a {personaje.get_nombre()}:")
        print(f"Origen: {personaje.get_origen()}")
        print(f"Descripción: {personaje.get_descripcion()}")

        respuesta = input("¿Los datos son correctos? (si/no): ").lower()

        if respuesta == "si":
            if personaje == self.monstruo_sin_permiso:
                print("¡No has identificado al monstruo sin permiso! ¡El monstruo ha escapado!")
                return False
            else:
                return True
        else:
            # Opción para llamar a los cazafantasmas
            cazafantasmas = input("¿Quieres llamar a los cazafantasmas para eliminar a este monstruo? (si/no): ").lower()

            if cazafantasmas == "si":
                if personaje == self.monstruo_sin_permiso:  # Comprobación si es el monstruo sin permiso
                    print("¡Correcto! Has llamado a los cazafantasmas y eliminado al monstruo sin permiso.")
                    print("¡Has salvado la noche de Halloween!")  # Mensaje de victoria
                    return True
                else:
                    print("¡Incorrecto! Has llamado a los cazafantasmas para un monstruo con permiso.")
                    print("¡El monstruo te ha matado en retaliación!")
                    return False
            else:
                print("No has encontrado al monstruo sin permiso.")
                return False


    def jugar(self):
        """Función principal del juego."""
        print("¡Bienvenido a Paper Please - Edición Halloween!")
        print("Tu misión es identificar al monstruo que no tiene permiso de ingresar.")

        self.seleccionar_personajes()
        self.mostrar_personajes()

        for personaje in self.personajes_seleccionados:
            resultado = self.cotejar_personaje(personaje)
            if not resultado:
                return


if __name__ == "__main__":
    juego = Juego()
    juego.jugar()

def heap_queue_largest(nums,n):
  largest_nums = hq.nlargest(n, nums)
  return largest_nums
```
